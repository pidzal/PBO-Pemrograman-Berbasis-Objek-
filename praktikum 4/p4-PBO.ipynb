```python
class Character:
    def __init__(self, nama, health, attack_power):
        self.__nama = nama 
        self.__health = health
        self.__attack_power = attack_power
        print(f"{self.__nama} bergabung ke dalam pertarungan!")
        
    def show_info(self):
        print("\n --- character info ---")
        print(f"nama {self.__nama}")
        print(f"health {self.__health}")
        print(f"attack power {self.__attack_power}")
        
    def attack(self, target):
        print(f"{self.__nama} menyerang {target.get_nama()}!")
        target.take_damage(self.__attack_power)
        
    def take_damage(self, damage):
        self.__health -= damage
        if self.__health <= 0 :
            print(f"{self.__nama} telah dikalahkan!")
            self.__health = 0 
        else:
            print(f"health {self.__nama} tersisa {self.__health}")
    
    def get_nama(self):
        return self.__nama
```

```python
class Warrior(Character):
    pass
```

```python
print("--- membuat objek dengan parent class ---")
aragorn = Character("Aragorn", 100, 15)

print("\n--- membuat objek dari child class ---")
gimil = Warrior("Gimil", 120, 20)
gimil.show_info()

print("\n--- pertarungan dimulai ---")
aragorn.attack(gimil)
gimil.attack(aragorn)
```

```python
class Warrior(Character):
    def __init__(self, nama, health, attack_power, armor):
        super().__init__(nama, health, attack_power)
    
        self.__armor = armor
        print(f"armor set to: {self.__armor}")
        
gimil = Warrior("gimil", 120, 20, 5)
gimil.show_info()
```
